# Step 2: Download HTML Documents

Downloads the full HTML page for each decision from the Bundesverfassungsgericht website.
Already-downloaded files are skipped so interrupted runs can be resumed safely.

**Input:** `data/2_list_clean/clean_list.csv` – list with `link` and `id` columns  
**Output:** `data/3_html_raw/{id}.html` – one HTML file per decision

```python
import os
import requests
import pandas as pd
```

```python
# Load the cleaned URL list and build a set of already-downloaded IDs to avoid re-downloading
data = pd.read_csv("data/2_list_clean/clean_list.csv", index_col=0)
downloaded_files = [f[:-5] for f in os.listdir("data/3_html_raw")]  # strip '.html' extension
```

```python
def download_html(link, id):
    # Fetch the decision page and save it as an HTML file
    response = requests.get(link)
    if response.status_code == 200:
        with open(f"data/3_html_raw/{id}.html", "w", encoding="utf-8") as f:
            f.write(response.text)
    else:
        print(f"Failed to download {id}: Status {response.status_code}")


# Iterate over all entries; skip files already present in data/3_html_raw/
for e in data.iterrows():
    print(e[0] + 1, end="\r")  # progress counter
    dec = e[1]
    if dec.id not in downloaded_files:
        download_html(dec.link, dec.id)
```
