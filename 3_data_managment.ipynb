# Step 3: Data Management

Organises the downloaded HTML files: filters to decisions from 2023–2025 and consolidates
them into a single working directory. For files that were manually corrected (hand-coded),
the corrected version takes precedence over the auto-downloaded one.

**Input:** `data/3_html_raw/` – all downloaded HTML files  
**Input:** `data/3_html_handcoding/` – manually corrected HTML files (take priority)  
**Output:** `data/4_html_raw_232425/` – 857 consolidated HTML files for 2023–2025

```python
import os
import re
import shutil
import pandas as pd
```

```python
# Build a DataFrame of all downloaded files and extract the decision year from the filename.
# Filename format: <2-char type prefix><4-digit year><...>.html
# Example: 'rk20241205_1bvr240624.html' -> year '2024'
data = pd.DataFrame({'files': os.listdir("data/3_html_raw")})
data['year'] = data.files.apply(lambda x: x[2:6])
data
```

```python
# Load list of manually corrected files; these override the auto-downloaded versions
handcoded_files = os.listdir("data/3_html_handcoding/")
handcoded_files
```

```python
# False id: case cs20250205_2bvc000725 ist from 2026!
data = data[data['files'] != "cs20250205_2bvc000725.html"]
```

```python
# Keep only decisions from 2023, 2024, and 2025 (858 files)
files232425 = data[data.year.isin(["2025", "2024", "2023"])].files.to_list()
len(files232425)
```

```python
# Copy files to the unified output directory.
# If a hand-coded (manually corrected) version exists, use it instead of the auto-download.
for e in files232425:
    if e in handcoded_files:
        shutil.copy("data/3_html_handcoding/" + e, "data/4_html_raw_232425/" + e)
    else:
        shutil.copy("data/3_html_raw/" + e, "data/4_html_raw_232425/" + e)
```

```python
# Verify all 858 files were copied successfully
len(os.listdir("data/4_html_raw_232425"))
```
